**Understanding the distribution of numerical features**

```python
def plot_hostogram(df, col_name):
  plt.figure(figsize=(10,6))
  sns.histplot(df[col_name], kde = True)
  plt.title(f'Distribution of {col_name}')

  # calculate the mean and median values
  col_mean = df[col_name].mean()
  col_median = df[col_name].median()

  # add vertical lines for mean and median
  plt.axvline(col_mean, color = 'red', linestyle = '--', label = 'Mean')
  plt.axvline(col_median, color = 'green', linestyle = '-', label = 'Median')

  plt.legend()
  plt.show()
```

#**4. Data Preprocessing**

**Label Encoding of Target column (Churn)**

```python
df['Churn'] = df['Churn'].replace({'No':0, 'Yes':1})
```

```python
df.head(5)
```

```python
print(df['Churn'].value_counts())
```

```python
df.head(2)
```

**Label Encoding of Categorical columns**

```python
#columns that data type is 'object' identified
object_columns = df.select_dtypes(include = 'object').columns
object_columns
```

```python
#initialize a dictionary to save the encoders
encoders = {}

#apply LabelEncoing to columns and store them in dictionary
for col in object_columns:
  encoder = LabelEncoder()
  df[col] = encoder.fit_transform(df[col])
  encoders[col] = encoder

#save the encoders to a pickle file
with open('encoders.pkl', 'wb') as f:
  pickle.dump(encoders, f)
```

```python
encoders
```

```python
df.head(3)
```

**Train Test Data Split**

```python
#Target & Features Split
x = df.drop(columns = ['Churn'])
y = df['Churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
print(y_train.shape)
```

```python
print(y_train.value_counts())
```

**Synthetic Minority Oversampling Technique (SMOTE)**

```python
smote = SMOTE(random_state = 42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
```

```python
print(y_train_smote.value_counts())
```

```python
print(x_train_smote.shape)
```

#**5. Model Training**

```python
models = {
    'Decision Tree' : DecisionTreeClassifier(random_state = 42),
    'Random Forest' : RandomForestClassifier(random_state = 42),
    'XGBoost' : XGBClassifier(random_state = 42)
}
```

```python
cv_scores = {}

#perform 5 fold cross validation for each model
for model_name, model in models.items():
  print(f'Training {model} with default parameters')
  scores = cross_val_score(model, x_train_smote, y_train_smote, cv = 5, scoring = 'accuracy')
  cv_scores[model_name] = scores
  print(f'{model_name} cross validation mean score: {np.mean(scores):.2f}')
  print('.'*70)
```

```python
cv_scores
```

Random Forest gives the highest accuracy compared to other models with default parameters

```python
rfc = RandomForestClassifier(random_state = 42)
rfc.fit(x_train_smote, y_train_smote)
```

```python
print(y_test.value_counts())
```

#**6. Model Evaluation**

```python
y_test_pred = rfc.predict(x_test)

print('Accuracy Score:\n', accuracy_score(y_test, y_test_pred))
print('.'*70)
print('Confusion Matrix:\n', confusion_matrix(y_test, y_test_pred))
print('.'*70)
print('Classification Report:\n', classification_report(y_test, y_test_pred))
```

```python
model_data = {
    'model' : rfc,
    'feature_names' : x.columns.tolist()
}

#save the trained model as a pickle file
with open('telco_churn_model.pkl', 'wb') as f:
  pickle.dump(model_data, f)
```

#**7. Load the saved model & predict churn**

```python
#Load the saved model & Features
with open('telco_churn_model.pkl', 'rb') as f:
  model_data = pickle.load(f)

loaded_model = model_data['model']
feature_names = model_data['feature_names']
```

```python
print(loaded_model)
```

```python
print(feature_names)
```

```python
input_data = {
    'gender': 'Female',
    'SeniorCitizen': 0,
    'Partner': 'Yes',
    'Dependents': 'No',
    'tenure': 1,
    'PhoneService': 'No',
    'MultipleLines': 'No phone service',
    'InternetService': 'DSL',
    'OnlineSecurity': 'No',
    'OnlineBackup': 'Yes',
    'DeviceProtection': 'No',
    'TechSupport': 'No',
    'StreamingTV': 'No',
    'StreamingMovies': 'No',
    'Contract': 'Month-to-month',
    'PaperlessBilling': 'Yes',
    'PaymentMethod': 'Electronic check',
    'MonthlyCharges': 29.85,
    'TotalCharges': 29.85
}

input_data_df = pd.DataFrame([input_data])

with open('encoders.pkl', 'rb') as f:
  encoders = pickle.load(f)

# encode the categorical features using saved encoders
for column, encoder in encoders.items():
  input_data_df[column] = encoder.transform(input_data_df[column])

#make a prediction
prediction = loaded_model.predict(input_data_df)
pred_prob = loaded_model.predict_proba(input_data_df)

print(prediction)

#Results
print(f"Prediction: {'Churn' if prediction == 1 else 'No Churn'}")
print(f'Prediction Probability: {pred_prob}')
```
